```python
# Import relevant libraries

import pandas as pd
from bokeh.layouts import column, row
from bokeh.models import ColumnDataSource, Select
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, curdoc
```

```python
# Read-in pre-processed data

df = pd.read_csv('/Users/loucap/Documents/GitWork/InteractiveGender/Data/cleaned_HLE.csv')
```

```python
# Prepare data - choose default qualification

df['selected_quali'] = df['No qualifications_grp']
df['selected_perc'] = df['Percentages_No qualifications']

# Let's set our source after we have sorted out column names out
source = ColumnDataSource(df)

# Create some tooltips to convey dp info

tool = [
    ("index", "$index"),
    ("(x,y)", "(@selected_quali{0.2f}, @selected_perc{0.2f})"),
    ("name", "@LA_name"),
]

# Create selection widget

options = ['No qualifications', 'Level 1', 'Level 2', 'Apprenticeships', 'Level 3', 'Level 4', 'Other']
select_qual = Select(title = "Highest qualification held:", value = 'No qualifications', options = options)

# Create second selection widget for colouring dps

option2 = ['Default', 'Region', 'Urban']
select_dps = Select(title = "Color By:", value = "None", options = option2)


# Create base figure to hold scatterplot

p = figure(title = "Relationship between highest qualification held and Non-response rate for LA's",
          x_axis_label = "Percentage of inhabitants with the highest qualification held",
          y_axis_label = "Non-response rate",
          tooltips = tool)

# Create scatterplot

p.scatter("selected_quali", "selected_perc", source = source, fill_alpha = 0.5, size = 10)


# Define callback to update data source for plot

def update_plot(attr, old, new):
    selected_qual = select_qual.value
    df['selected_quali'] = df[f'{selected_qual}_grp']
    df['selected_perc'] = df[f'Percentages_{selected_qual}']

    for region, scatter in scatter_plots_per_region.items():
        b = df[df.region == region]
        scatter.data_source.data = b

    source.data = df


    
# Call to update_plot function

select_qual.on_change('value', update_plot)

# Let's create our other graphs

# Region

p1 = figure(title = "Relationship between highest qualification held and Non-response rate for LA's",
           x_axis_label = "Percentage of inhabitants with the highest qualification held",
           y_axis_label = "Non-response rate",
           tooltips = tool)

okabe_ito = colorblind_palette = ['#E69F00', '#56B4E9', '#009E73', '#F0E442', '#0072B2', '#D55E00', '#CC79A7', '#000000', '#006400', '#800000']
# Create a dictionary to store scatter plots per region
scatter_plots_per_region = {}

for (idx, region) in enumerate(df.region.unique()):
    b = df[df.region == region]
    color = okabe_ito[idx % len(okabe_ito)]  # cycle through colors
    scatter = p1.circle(x="selected_quali", y="selected_perc", size=10, alpha=0.5, color=color,
              legend_label=region, muted_color=color, muted_alpha=0.1, source=ColumnDataSource(b))
    scatter_plots_per_region[region] = scatter
    
p1.legend.location = "bottom_right"
p1.legend.click_policy = "hide"
p1.legend.title = "Regions"

# Define second callback function
def update_scatterplots(attr, old, new):
    if select_dps.value == "Default":
        p.visible = True
        p1.visible = False
    if select_dps.value == "Region":
        p.visible = False
        p1.visible = True
        
# Set initial visibility
p.visible = True
p1.visible = False

# Call to second update function
select_dps.on_change('value', update_scatterplots)

# Create layout
layout = row(select_qual, select_dps, p, p1)

# show(layout)
output_notebook()

# Render to Bokeh server

curdoc().add_root(layout)
```
