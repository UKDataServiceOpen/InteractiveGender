```python
import pandas as pd
```

```python
rel = pd.read_excel('/Users/loucap/Documents/GitWork/InteractiveGender/Data/religion_gi.xlsx')
```

```python
rel.rename(columns={'Lower tier local authorities Code':'LA_code', 'Lower tier local authorities':'LA_name', 'Gender identity (4 categories) Code': 'GI_code', 'Gender identity (4 categories)': 'GI_categories', 'Religion (10 categories) Code': 'Religion_code', 'Religion (10 categories)': 'Religion_categories'}, inplace = True)
```

```python
rel.head()
```

```python
rel = rel[(rel.GI_code != -8) & (rel.Religion_code != -8)]
```

```python
sub = rel[(rel.GI_code != -8) & (rel.Religion_code != -8)]
```

```python
diff = sub[sub.GI_categories == 'Gender identity different from sex registered at birth']
```

```python
b = diff.groupby('Religion_categories')['Observation'].sum().reset_index()
```

```python
b
```

```python
x = rel.groupby('Religion_categories')['Observation'].sum().reset_index()
```

```python
x
```

```python
x['Percentages'] = round(b['Observation'] / x['Observation'] * 100, 2)
```

```python
x.sort_values(by = 'Percentages', ascending = False)
```
