## Percentages 

```python
df['Percentages'] = ''

for i in df.LA_code.unique():
    b = df[df.LA_code == i]
    percent = (b['Observation'] / b['Observation'].sum() * 100).round(2)
    df.loc[b.index, 'Percentages'] = percent
df.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define colors for each GI_categories value
category_colors = sns.color_palette("husl", len(df.GI_categories.unique()))
```

```python
df.GI_categories.unique()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define custom colors for each GI_categories value
category_color_mapping = {
    'Gender identity the same as sex registered at birth': 'yellow',
    'Gender identity different from sex registered at birth but no specific identity given': 'grey',
    'Trans woman': 'pink',
    'Trans man': 'blue',
    'All other gender identities': 'purple',
    'All other gender identities': 'black',
    'Not answered': 'red'
    # Add more categories and colors as needed
}

for category in df.GI_categories.unique():
    category_data = df[df['GI_categories'] == category]
    top_10_LA = category_data.sort_values('Percentages', ascending=False).head(10)

    plt.figure(figsize=(12, 6))

    # Pass the custom color for the current GI_categories value
    sns.barplot(data=top_10_LA, x='LA_name', y='Percentages', color=category_color_mapping[category])

    plt.xticks(rotation=45)
    plt.xlabel('Local Authorities')
    plt.ylabel('Percentages')
    plt.title(f'Top 10 Local Authorities with the Highest Observations of {category}')
    plt.show()
```

```python
c = b.sort_values('Percentages', ascending = False)
```

```python
# Group by 'Religion Code'
grouped_c = c.groupby('Religion (10 categories)', as_index=False)

# Calculate the sum of 'Percentages' for each group
summed_percentages = grouped_c.agg({'Percentages': 'sum'})

# Calculate the total percentage for each religion
total_percentages = summed_percentages['Percentages'] / summed_percentages['Percentages'].sum() * 100

# Add the total percentages to the resulting DataFrame
summed_percentages['Total Percentages'] = total_percentages

summed_percentages.sort_values('Total Percentages', ascending = False)
```
